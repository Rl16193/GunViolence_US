```python
# Initial imports
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
def remove_numbers(data):
    data1 = data.replace(r'0-9[^A-Za-z]', '')
    return data1
```

```python
# Downloading contents of the web page
url = "https://en.m.wikipedia.org/wiki/List_of_U.S._states_and_territories_by_income"
data = requests.get(url).text
```

```python
# Creating BeautifulSoup object
soup = bs(data, 'html.parser')
```

```python
print(soup.prettify())
```

```python
column_names = ["Rank", "States", "Per capita income (2019)", "Personal per capita income (2020)", 
                "Of which disposable personal per capita income (2020)", "Median household income (2019)", 
                "Median family income (2019)", "Population (April 1, 2020)", "Number of households (2019)", 
                "Number of families (2019)"]
tabdata = []
```

```python
i = 0
for table in soup.find_all('table'):
    
    i += 1 #Iterate to the thrid table as that is the one we need
    if (i ==3):
        
        # Collecting Ddata
        for row in table.tbody.find_all('tr'):
            data = []
            # Find all data for each column
            for columns in row.find_all('td'):
                data.append(columns.text.strip())
            print(data)
            tabdata.append(data)
```

```python
df_popu_income = pd.DataFrame(tabdata, columns = column_names)
```

```python
# Obtain the income data
i = 0
tabdata = []
for table in soup.find_all('table'):
    
    i += 1 #Iterate to the thrid table as that is the one we need
    if (i ==1):
        
        # Collecting Ddata
        for row in table.tbody.find_all('tr'):
            data= []
            data.append(row.contents[2].text.strip())
            data.append(row.contents[24].text.strip())
            tabdata.append(data)
```

```python
incomegrowth_df = pd.DataFrame(tabdata, columns = ['States','Growth in Income over 10 years(%)'])
incomegrowth_df.head()
```

```python
# Remove Fist row

incomegrowth_df.drop(0, inplace = True)
```

```python
incomegrowth_df.head()
```

```python
# Clean the DataFrame

df_popu_income.drop(columns = ['Rank', 'Per capita income (2019)','Personal per capita income (2020)',
                               'Of which disposable personal per capita income (2020)'], inplace = True)
df_popu_income.dropna(how = "any", inplace = True)
```

```python
# Downloading contents of the web page for INCOME INEQUALITY
url = "https://en.wikipedia.org/wiki/List_of_U.S._states_and_territories_by_income_inequality"
data_url= requests.get(url).text
```

```python
# Creating BeautifulSoup object
soup = bs(data_url, 'html.parser')
```

```python
# Get the table from the wikipedia page

i=0
tabdata = []
data = []
for table in soup.find_all('table'):
    
    i += 1
   
    if (i==2): #Iterate to the second table to get % annual change in income inequality
        
        # Collecting Data
        for tr in table.tbody.find_all('tr'):
            data= []
            # Find all data for each column
            for columns in tr.find_all('td'):
                data.append(columns.text.strip())
            tabdata.append(data)
```

```python
#Create Inequality Dataset

inequality_df = pd.DataFrame(tabdata, columns = ['#', 'States', 'Inequality coefficient (2019)', 'Inequality coefficient (2015-2019)'])
```

```python
#Clean DataFrame

inequality_df.drop(columns = '#', inplace = True)
inequality_df.drop(0, inplace = True)
```

```python
inequality_df.head()
```

```python
# Downloading contents of the web page for unemployment rate
url = "https://en.m.wikipedia.org/wiki/List_of_U.S._states_and_territories_by_unemployment_rate"
data_url= requests.get(url).text
```

```python
# Creating BeautifulSoup object
soup = bs(data_url, 'html.parser')
```

```python
 # Collecting Data

tabdata = []
data = []

for table in soup.find_all('table'):
    for tr in table.tbody.find_all('tr'):
        data= []
        data.append(tr.contents[4].text.strip())
        data.append(tr.contents[6].text.strip())
        tabdata.append(data)
```

```python
# Create Unemployment Dataset

unemployment_df = pd.DataFrame(tabdata, columns = ['States', 'Unemployment Rates'])
```

```python
# Clean DataFrame

unemployment_df.drop(0, inplace = True)
unemployment_df.reset_index(drop = True)
unemployment_df['States'] = remove_numbers(unemployment_df['States'])
```

```python
unemployment_df.head()
```

```python
# Downloading contents of the web page for poverty rate
url = "https://en.wikipedia.org/wiki/List_of_U.S._states_and_territories_by_poverty_rate"
data_url= requests.get(url).text
```

```python
# Creating BeautifulSoup object
soup = bs(data_url, 'html.parser')
```

```python
# Collect Data from Webpage

tabdata = []
for table in soup.find_all('table'):
    if 'mw-datatable' in table.get('class'):
        for tr in table.tbody.find_all('tr'):
            data= []
            data.append(tr.contents[1].text.strip())
            data.append(tr.contents[7].text.strip())
            tabdata.append(data)
```

```python
# Create Poverty Dataset

poverty_df = pd.DataFrame(tabdata, columns = ['States', 'Poverty Rate'])
```

```python
# Clean Data Frame

poverty_df.drop([0,1], inplace = True)
poverty_df.reset_index(drop = True, inplace = True)
poverty_df['States'] = remove_numbers(poverty_df['States'])
```

```python
poverty_df.head()
```

```python
# Downloading contents of the web page for poverty rate
url = "https://educationdata.org/public-education-spending-statistics"
data_url= requests.get(url).text
```

```python
# Creating BeautifulSoup object
soup = bs(data_url, 'html.parser')
```

```python
# Collect Data from Webpage

i=0
tabdata = []

for table in soup.find_all('table'):
    
    i += 1
    if (i ==3): #Iterate to the thrid table as that is the one we need
        
        # Collecting Data
        for row in table.tbody.find_all('tr'):
            data = []
            # Find all data for each column
            for columns in row.find_all('td'):
                data.append(columns.text.strip())
            tabdata.append(data)
```

```python
# Create School Funding Dataset

funding_df = pd.DataFrame (tabdata, columns = ['States','Federal Funding','State Funding'])
```

```python
funding_df.head()
```

```python
# Downloading contents of the web page for unemployment rate
url = "https://www.statsamerica.org/sip/rank_list.aspx?rank_label=censgovtre_exp_1_c&item_in=040"
data_url= requests.get(url).text
```

```python
# Collect Data from Webpage

i=0
tabdata = []

for table in soup.find_all('table'):
    
    for tr in table.find_all('tr'):
        data= []
        data.append(tr.contents[1].text.strip())
        data.append(tr.contents[3].text.strip())
        tabdata.append(data)
```

```python
# Create State Expenditure

social_spend_df = pd.DataFrame (tabdata, columns = ['States','Per Capita Expense on Social Welfare'])
```

```python
# Clean Data
social_spend_df.drop(0, inplace = True)
```

```python
# View the Data
social_spend_df.head(10)
```

```python
df1 = pd.merge(df_popu_income, incomegrowth_df, how = 'inner')
df1 = pd.merge(df1, inequality_df, how = 'inner')
df1 = pd.merge(df1,unemployment_df,how = 'inner')
df1 = pd.merge(df1, poverty_df, how = 'inner')
df1 = pd.merge(df1, funding_df, how = 'inner')
df1 = pd.merge(df1, social_spend_df, how = 'inner')
```
